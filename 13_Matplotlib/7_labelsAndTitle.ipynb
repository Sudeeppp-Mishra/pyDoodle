# Matplotlib Labels and Title

## Create Labels for a Plot

With Pyplot, you cna use the `xlabel()` and `ylabel()` functions to set a label for the x- and y-axis.


```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([3, 10])
ypoints = np.array([10, 20])

plt.plot(xpoints, ypoints)

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.show()
```

## Create a Title for a Plot

With Pyplot, you can use the `title()` funciton to set a title for the plot.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([9, 10, 11, 1,2, 13])

plt.plot(x,y)

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.title("Plotting")

plt.show()
```

## Set Font Properties for Title and Labels

You can use the `fontdict` parameter in `xlabel()`, `ylabel()`, and `title()` to set font properties for the title and labels.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([10, 20])
y = np.array([100, 200])

font1 = {'family': 'serif', 'color': 'blue', 'size': 20}
font2 = {'family': 'times new roman', 'color': 'red', 'size': 15}

plt.title('Graph', fontdict = font1)
plt.xlabel('X-Axis', fontdict = font2)
plt.ylabel('Y-Axis', fontdict = font2)

plt.plot(x,y)
plt.show()
```

## Position the Title

You can use the `loc` parameter in `title()` to position the title

Legal values are: 'left', 'right', and 'center'. Default value is 'center'.

```python
plt.title('Graph', loc = 'right')

plt.plot(x,y)
plt.show()
```
